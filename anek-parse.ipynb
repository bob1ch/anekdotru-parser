```python
con = sqlite3.connect("i_hate_jokes.db")
cur = con.cursor()
```

```python
cur.execute("DROP TABLE IF EXISTS jokes")
cur.execute("""CREATE TABLE jokes (
                                    id INTEGER PRIMARY KEY AUTOINCREMENT,
                                    Anekdot TEXT,
                                    Author TEXT,
                                    Rating INTEGER,
                                    URL_Anek TEXT,
                                    URL_Author TIME,
                                    Date_Anek TEXT)""")
```
